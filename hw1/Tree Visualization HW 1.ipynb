# Реализация

## Считывание и отрисовка дерева посредством класса **GraphMLParser**

```python
graph_parser = GraphMLParser()

def read_xml_graph(filename):
  graph_xml = graph_parser.parse(filename)
  graph_xml.show()
  nodes = graph_xml.nodes()
  root_node = set(map(lambda x: x.id, nodes))
  graph = dict.fromkeys(root_node, [])

  graph_edges = graph_xml.edges()
  for edge in graph_edges:
    graph[edge.node1.id] = graph[edge.node1.id] + [edge.node2.id]
    root_node.remove(edge.node2.id)

  root_node = list(root_node)[0]
  return (root_node, graph)
```

## Реализация метода

```python
def draw(root_node, graph):
  global current_graph
  current_graph = graph
  coordinates, bounds = inner_draw(root_node,0)
  return coordinates

def inner_draw(vertex, depth):
  nodes_coord = dict()

  if not current_graph[vertex]:
    bounds = np.array([[0,0]])
    nodes_coord[vertex] = (0, depth)
    return nodes_coord, bounds

  if len(current_graph[vertex]) == 1:
    nodes_coord, left_bound = inner_draw(current_graph[vertex][0], depth - 1)
    bounds = np.array([ np.array([0, 0]), *left_bound])
    nodes_coord[vertex] = (0, depth)
    return nodes_coord, bounds

  nodes_coord, left_bound = inner_draw(current_graph[vertex][0], depth - 1)
  right_coord, right_bound = inner_draw(current_graph[vertex][1], depth - 1)

  min_depth = min(len(left_bound), len(right_bound))
  max_depth = max(len(left_bound), len(right_bound))

  copy_left_bound = left_bound
  copy_right_bound = right_bound

  shift = 0
  for i in range(min_depth):
    cur_shift = left_bound[i][1] + 1 - right_bound[i][0]
    if cur_shift > shift:
      shift = cur_shift

  right_bound = list(map(lambda x: list(map(lambda y: y + shift, x)), right_bound))

  copy_right_bound = right_bound

  bounds = np.zeros((max_depth, 2))

  for i in range(max_depth):
    if i >= len(left_bound):
      left = copy_right_bound[i][0]
      right = copy_right_bound[i][1]
    elif i >= len(right_bound):
      left = copy_left_bound[i][0]
      right = copy_left_bound[i][1]
    else:
      left = min(copy_left_bound[i][0], copy_right_bound[i][0])
      right = max(copy_left_bound[i][1], copy_right_bound[i][1])

    bounds[i][0] = left
    bounds[i][1] = right

  nodes_coord[vertex] = (0, depth)
  bounds = np.array([np.array([0, 0]), *bounds])

  shift_right_coord = dict()
  for (node, coord) in right_coord.items():
    shift_right_coord[node] = (coord[0] + shift, coord[1])

  nodes_coord.update(shift_right_coord)
  return nodes_coord, bounds
```

## Отрисовка графа с помощью **MatPlotlib**

```python
def show(graph, coordinates):
  fig, ax = plt.subplots()

  ax.plot([i[0] for i in coordinates.values()], [i[1] for i in coordinates.values()], 'o', color='black');

  for (node, children) in graph.items():
    parent_coord = coordinates[node]
    for child in children:
      child_coord = coordinates[child]
      x_values = (parent_coord[0], child_coord[0])
      y_values = (parent_coord[1], child_coord[1])
      ax.plot(x_values, y_values, color='blue')

  ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
  ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
  ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
  ax.yaxis.set_minor_locator(ticker.MultipleLocator(2))

  ax.grid("True")
  fig.set_figwidth(14)
  fig.set_figheight(10)

  plt.show()
```
